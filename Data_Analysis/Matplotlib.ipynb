# Matplotlib数据可视化基础

## 绘图基本语法与常用参数

### pyplot

- 创建画布与创建子图

|函数名|函数说明|
|:--:|:--:|
|`pyplot.figure`|创建一个空白画布，可以指定大小、像素|
|`figure.add_subplot`|创建并选中子图，可以指定子图行列数，与选中图片编号|

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(4,4))
```

- 添加画布内容

|函数名|函数说明|
|:--:|:--:|
|plt.title|添加标题|
|plt.xlabel|添加x轴名称|
|plt.ylabel|添加y轴名称|
|plt.xlim|指定x轴范围|
|plt.ylim|指定y轴范围|
|plt.xticks|指定x轴刻度的数目和取值|
|plt.yticks|指定y轴刻度的数目和取值|
|plt.legend|指定当前图形的图例|

```python
x = np.arange(10)
plt.title('sin  cos')
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.legend(['sin','cos'])
```

- 保存与展示图形

|函数名|函数说明|
|:--:|:--:|
|plt.savafig|保存绘制的图片|
|plt.show|在本机展示图片|

## 分析特征间关系

### 散点图

#### scatter函数

`matplotlib.pyplot.scatter(x, y, s = None, c = None, marker = None, alpha = None, **kwargs)`

|参数名称|说明|
|:--:|:--:|
|x，y|接收array，表示x轴和y轴数据|
|s|接收数组或者一维array，指定点的大小|
|c|同上，指定点的颜色|
|marker|接收特定string，表示绘制点类型|
|alpha|接收0-1的小数，表示点的透明度|

```python
data = np.load('./data/国民经济核算季度数据.npz', allow_pickle= True)
```

```python
data.files
```

```python
plt.scatter(range(69), data['values'][: ,2])
plt.show()
```

```python
for i in [3, 4, 5]:
    plt.scatter(range(69), data['values'][ : , i])
plt.legend(['1','2','3'])
plt.show()
```

## 折线图

### plot函数

`matplotlib.pyplot.plot(*args, **kwargs)`

|参数名|参数说明|
|:--:|:--:|
|x, y|接收array。表示x轴和y轴数据|
|color|接收特定string。指定线条颜色|
|linestyle|接收特定string。指定线条类型，默认‘-’|
|marker|接收特定string。表示绘制点类型|
|alpha|接收0-1小数。表示点透明度|

#### 颜色缩写

|颜色缩写|颜色|颜色缩写|颜色|
|:--:|:--:|:--:|:--:|
|b|蓝色|m|品红|
|g|绿色|y|黄色|
|r|红色|k|黑色|
|c|青色|w|白色|

```python
d = ['o','*','D']
for j,i in enumerate([3,4,5]):
    plt.plot(range(69),data['values'][:,i], marker = d[j])
plt.legend(['1','2','3'])
plt.show()
```

## 分析特征内部数据分布与分散状况

### 直方图

#### bar函数

`matplotlib.pyplot.bar(left, height, width = 0.8, bottom = None, hold = None, data = None,**kwargs)`

|参数名称|参数说明|
|:--:|:--:|
|left|接收array。表示x轴数据|
|height|接收array。表示x轴所代表数据的数量|
|width|接收0-1的float。指定直方图宽度，默认0.8|
|color|接收特定string或颜色字符串的array。表示直方图颜色|

```python
num = data['values'][68,3:6]
```

```python
plt.bar(range(len(num)), num)
plt.xticks(range(len(num)),[1,2,3])
plt.show()
```

### 饼图

#### pie函数

`matplotlib.pyplot.pie(x, explode = None, labels = None, color = None, autopct = None, pctdistance = 0.6, shadow = False, labeldistance = 1.1, startangle = None, radius = None,...)`

|参数名称|说明|参数名称|说明|
|:--:|:--:|:--:|:--:|
|x|接收array。表示用于pie的数据|autopct|接收特定string。指定数值的显示方式|
|explode|接收array。表示指定项离饼图圆心为n个半径|pctdistance|接收float。指定每一项的比例和距离饼心n个半径，默认 0.6|
|labels|接收array。指定每一项名称|labeldistance|接收float。指定每一项名称和距离饼心n个半径。默认 1.1|
|color|接收特定string或者包含特定颜色字符串的array。表示饼图颜色|radius|接收float。表示饼图半径，默认 1|

```python
plt.figure(figsize=(4,4)) #正方形画布
plt.pie(num, autopct='%.2f %%', explode=[0.1,0,0], labels=['1','2','3'], labeldistance= 1.4)
plt.show()
```

### 箱线图

- 分散程度差异

- 五个统计量（最小值，下四分位数，中位数，上四分位数，最大值，#异常值）

#### boxplot函数

`matplotlib.pyplot.boxplot(x, notch = None, sym = None, vert = None, whis = None, positions = None,widths = None,patch_artist = None, meanline = None, labels = None,...)`

|参数名称|说明|参数名称|说明|
|:--:|:--:|:--:|:--:|
|x|接收array。表示箱线图数据|positions|接收array。表示图形位置|
|notch|接收bool。表示中间箱体是否有缺口|widths|接收scalar或array。表示箱体宽度|
|sym|接收特定string。指定异常点形状|labels|接收array。指定每个箱线图的标签|
|vert|接收bool。表示图形横向或纵向|meanline|接收bool。表示是否显示均值线|

```python
num = (list(data['values'][:,3]), list(data['values'][:,4]), list(data['values'][:,5]))

plt.boxplot(num)
plt.show()
```
